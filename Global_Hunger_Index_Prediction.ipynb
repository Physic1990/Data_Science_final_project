Prediction Model for different countries


```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import plotly.graph_objects as go
import warnings

# Load the dataset with specified encoding
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Global hunger index.csv', encoding='ISO-8859-1')

# Data Preprocessing
# Remove unnecessary columns
data = data.drop(['Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11'], axis=1)

# Convert numeric columns to the appropriate data type
for year in ['2000', '2007', '2014', '2022']:
    data[year] = pd.to_numeric(data[year], errors='coerce')

# Transpose the DataFrame to set 'Country' as index
data = data.set_index('Country').T

# Suppress ARIMA warnings
warnings.filterwarnings("ignore")

# Dictionary to store predictions for each country
predictions = {}

# Iterate over each country
for country in data.columns:
    try:
        # Copy the original data
        df = data[[country]].copy()

        # Convert the data type of the country column to numeric
        df[country] = pd.to_numeric(df[country], errors='coerce')

        # Train the ARIMA model
        model = ARIMA(df[country], order=(5, 1, 0))  # Example order, you may need to tune this
        model_fit = model.fit()

        # Make predictions for the next 50 years
        forecast = model_fit.forecast(steps=50)

        # Store the predictions
        predictions[country] = forecast
    except Exception as e:
        print(f"Error occurred while processing {country}: {e}")

# Create a figure
fig = go.Figure()

# Add traces for each country's predicted values
for country, forecast in predictions.items():
    fig.add_trace(go.Scatter(x=list(range(2023, 2073)), y=forecast, mode='lines', name=country))

# Update layout
fig.update_layout(title='Predicted Hunger Index for Each Country Over the Next 50 Years',
                  xaxis_title='Year',
                  yaxis_title='Hunger Index')

# Show the plot
fig.show()

# Create individual line charts for each country
for country, forecast in predictions.items():
    # Create a figure
    fig = go.Figure()

    # Add trace for historical data
    fig.add_trace(go.Scatter(x=data.index, y=data[country], mode='lines', name='Historical Data'))

    # Add trace for predicted values
    fig.add_trace(go.Scatter(x=list(range(2023, 2073)), y=forecast, mode='lines', name='Predicted Data'))

    # Update layout
    fig.update_layout(title=f'Hunger Index for {country} (Historical and Predicted)',
                      xaxis_title='Year',
                      yaxis_title='Hunger Index')

    # Show the plot
    fig.show()
```

LSTM neural network

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import plotly.graph_objects as go

# Load the dataset with specified encoding
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/more_years.csv', encoding='ISO-8859-1')

# Extract the hunger index data for a specific country (replace 'CountryName' with the actual country name)
country_data = data.loc[data['Country'] == 'Afghanistan', ['2007', '2008', '2014', '2015', '2022', '2023']]
country_data = country_data.squeeze()  # Convert DataFrame to Series

# Preprocess the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(np.array(country_data).reshape(-1, 1))

def create_dataset(data, time_steps):
    X, y = [], []
    for i in range(len(data) - time_steps + 1):
        X.append(data[i:(i + time_steps), 0])
        y.append(data[i + time_steps - 1, 0])  # Adjusted index to capture correct label
    X, y = np.array(X), np.array(y)
    print("Number of samples:", len(X))
    print("Shape of X:", X.shape)
    print("Shape of y:", y.shape)
    return X, y

# Set the number of time steps (e.g., 5 years)
time_steps = 5

# Create the dataset for training the LSTM model
X, y = create_dataset(scaled_data, time_steps)

# Reshape data for LSTM input (samples, time steps, features)
X = X.reshape(X.shape[0], X.shape[1], 1)

# Split the data into training and testing sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

print("Training data shape:", X_train.shape, y_train.shape)
print("Testing data shape:", X_test.shape, y_test.shape)

# Build the LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dense(units=1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=100, batch_size=32, verbose=1)

# Make predictions
train_predictions = model.predict(X_train)
test_predictions = model.predict(X_test)

# Inverse transform predictions to original scale
train_predictions = scaler.inverse_transform(train_predictions)
test_predictions = scaler.inverse_transform(test_predictions)

# Print predictions for the future 10 years
future_data = scaled_data[-time_steps:].reshape(1, time_steps, 1)
future_predictions = []

for i in range(10):
    prediction = model.predict(future_data)
    future_predictions.append(prediction.flatten()[0])
    future_data = np.append(future_data[:, 1:, :], prediction.reshape(1, 1, 1), axis=1)

future_predictions = scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))
print("Future predictions for the next 10 years:")
print(future_predictions.flatten())

# Plotting results
fig = go.Figure()

# Plot training data
fig.add_trace(go.Scatter(x=country_data.index[:train_size], y=country_data[:train_size], mode='lines', name='Training Data'))

# Plot training predictions
train_prediction_indices = range(time_steps, train_size + 1)
fig.add_trace(go.Scatter(x=country_data.index[train_prediction_indices], y=train_predictions.flatten(), mode='lines', name='Training Predictions'))

# Plot testing data
fig.add_trace(go.Scatter(x=country_data.index[train_size+time_steps:], y=country_data[train_size+time_steps:], mode='lines', name='Testing Data'))

# Plot testing predictions
test_prediction_indices = range(train_size + time_steps, len(country_data))
fig.add_trace(go.Scatter(x=country_data.index[test_prediction_indices], y=test_predictions.flatten(), mode='lines', name='Testing Predictions'))

# Create a list of years from 2024 to 2033
future_years = list(range(2024, 2034))

# Plot future predictions
fig.add_trace(go.Scatter(x=future_years, y=future_predictions.flatten(), mode='lines', name='Future Predictions'))

fig.update_layout(title='Hunger Index Forecast using LSTM',
                  xaxis_title='Year',
                  yaxis_title='Hunger Index')

fig.show()
```
